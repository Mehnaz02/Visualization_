```python
# Plot the top product types based on the number of articles
plt.figure(figsize=(14, 7))
product_type_counts = art['product_type_name'].value_counts().head(10)
sns.barplot(x=product_type_counts.index, y=product_type_counts.values, palette='viridis')
plt.title('Top 10 Product Types by Number of Articles')
plt.xlabel('Product Type')
plt.ylabel('Number of Articles')
plt.xticks(rotation=45)
plt.show()
```

```python
#Colour Group Name: Analyzing the distribution of products by color groups can reveal trends in color
#preferences among customers and help identify popular color palettes.

colour_group_counts = art['colour_group_name'].value_counts()
# Plotting the distribution of products by colour group
plt.figure(figsize=(17, 5))
colour_group_counts.plot(kind='bar', color='pink')
plt.title('Distribution of Products by Colour Group')
plt.xlabel('Colour Group')
plt.ylabel('Number of Products')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
department_counts = art['department_name'].value_counts()

# Plotting the distribution of products by department
plt.figure(figsize=(30, 6))
department_counts.plot(kind='bar', color='lightcoral')
plt.title('Distribution of Products by Department')
plt.xlabel('Department')
plt.ylabel('Number of Products')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
garment_group_counts = art['garment_group_name'].value_counts()

# Plotting the distribution of products by garment group
plt.figure(figsize=(10, 6))
garment_group_counts.plot(kind='bar', color='lightblue')
plt.title('Distribution of Products by Garment Group')
plt.xlabel('Garment Group')
plt.ylabel('Number of Products')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
cust.duplicated().sum()
```

#Visualization

```python
# Histogram of customer ages
plt.figure(figsize=(15, 5))
sns.histplot(data=cust, x='age', kde=True)
plt.title('Distribution of Customer Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Violin plot of age by fashion news frequency and active status
plt.figure(figsize=(12, 6))
sns.violinplot(data=cust, x='fashion_news_frequency', y='age', hue='Active', split=True)
plt.title('Age Distribution by Fashion News Frequency and Active Status')
plt.xlabel('Fashion News Frequency')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=cust, x='Active')
plt.title('Distribution of Active and Inactive Customers')
plt.xlabel('Active Status')
plt.ylabel('Count')
plt.show()
```

```python
cust['age'] = cust['age'].astype(str)
cust['fashion_news_frequency'] = cust['fashion_news_frequency'].astype(str)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=cust, x='age', y='fashion_news_frequency')
plt.title('Relationship between Age and Fashion News Frequency')
plt.xlabel('Age')
plt.ylabel('Fashion News Frequency')
plt.show()
```

```python
freq_map = {0: 'Never', 1: 'Monthly', 2: 'Weekly', 3: 'Daily'}

# Replace numerical values with string labels
cust['fashion_news_frequency'] = cust['fashion_news_frequency'].map(freq_map)

# Scatter plot of age vs. fashion news frequency
plt.figure(figsize=(8, 6))
sns.scatterplot(data=cust, x='age', y='fashion_news_frequency')
plt.title('Relationship between Age and Fashion News Frequency')
plt.xlabel('Age')
plt.ylabel('Fashion News Frequency')
plt.show()
```

```python
# Count plot of postal codes
plt.figure(figsize=(12, 6))
sns.countplot(data=cust, x='postal_code')
plt.xticks(rotation=90)
plt.title('Distribution of Postal Codes')
plt.xlabel('Postal Code')
plt.ylabel('Count')
plt.show()
```
